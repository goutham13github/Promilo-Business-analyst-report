```python
import pandas as pd
df = pd.read_excel("dataset.xlsx")
df
```

```python
df.dtypes
```

```python
# removing null values
df=df[df["Users"]!=0].dropna()
df=df[df["Users"]>=20].dropna()
df=df[df["Users"]!=36].dropna()
```

```python
df=df.drop("Total revenue",axis=1)
df
```

```python
import matplotlib.pyplot as plt
x=df["Country"]
y=df["Users"]
plt.bar(x,y)
plt.xlabel("Country")
plt.ylabel("Users")
plt.title("Users in different countries")
plt.xticks(rotation=45)
plt.show()
```
